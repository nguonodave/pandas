## Making changes to the data

```python
# df['Weight/Height'] = df['Weight']/df['Height'] # adding a column with the given name
# another way of doing the above is as below
df['Weight/Height'] = df.iloc[:, 2:4].sum(axis=1) # semicolon means all the rows, 2 and 4 means from the column at index 2 to index 3(excluding index 4 column), axis=1 means going horizontally(adding through rows)

df.head()
```

```python
# switching a column to a particular position
cols = list(df.columns) # this will get all the positions of the columns and store them in a list
df = df[cols[0:4] + [cols[-2]] + cols[5:16] + [cols[-1]]]
df.head(7)
```

```python
df = df.drop(columns=["Weight/Height"]) # dropping(deleting a column)

df.head()
```

## Conditional changes/renaming entries

```python
df.loc[df['FAVC'] == 'yes', 'FAVC'] = 'yeahp' # this means that if entries in column 'FAVC' is yes then change it to 'yeahp'
df
```

```python
df.loc[df['FAVC'] == 'yeahp', 'Gender'] = 'male'  # this means that if entries in column 'FAVC' is yeahp then change the corresponding entries in column 'Gender' to 'male'
df
```

```python
# doing the same as above but for multiple columns(modifying multiple columns on a certain condition)
# df.loc[df['Age'] > 18, ['FAVC', 'FCVC']] = 'TEST' # if age is over 18 then set entries in columns 'FAVC' and 'FCVC' to TEST
df.loc[df['Age'] > 18, ['FAVC', 'FCVC']] = ['TEST 1', 'TEST 2'] # doing the same as above but modifying multiple columns individually
df
```
